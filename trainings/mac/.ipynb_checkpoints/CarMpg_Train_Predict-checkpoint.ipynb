```python
import h2o
```

```python
url = "https://raw.githubusercontent.com/Avkash/mldl/master/data/auto_mpg.csv"
```

```python
h2o.init()
```

```python
df = h2o.import_file(url)
```

```python
df
```

```python
df.shape
```

```python
df.col_names
```

```python
y = "mpg"
```

```python
x = list(df.columns)
```

```python
x.remove(y)
```

```python
x
```

```python
from h2o.estimators.random_forest import H2ORandomForestEstimator
```

```python
rf_fit1 = H2ORandomForestEstimator(model_id='model_mpg', seed=1, nfolds=5)
```

```python
rf_fit1.train(x=x, y=y, training_frame=df)
```

```python
rf_fit1
```

```python
rf_perf1 = rf_fit1.model_performance(df)
```

```python
rf_fit1
```

```python
pred = df[1,:6]
```

```python
pred['cylinders'] = 16
pred['horsepower'] = 400
pred['weight'] = 8000
```

```python
pred
```

```python
rf_fit1.predict(pred)
```
