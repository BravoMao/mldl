```python
import h2o
import pandas as pd
```

```python
h2o.init()
```

# Importing Dataset

```python
df = h2o.import_file('/Users/avkashchauhan/learn/customers/mojo-cpp/airlines_train.csv')
```

```python
df.shape
```

```python
df.col_names
```

```python
y = "IsDepDelayed"
```

```python
x = df.col_names
x.remove(y)
print(x)
```

# Building GBM model using H2O

```python
from h2o.estimators.gbm import H2OGradientBoostingEstimator
```

```python
gbm_model = H2OGradientBoostingEstimator()
```

```python
gbm_model.train(x = x, y = y, training_frame=df)
```

```python
print(gbm_model)
```

```python
print("Total trees in the model : " + str(gbm_model.default_params['ntrees']))
```

```python
scoring_hist = gbm_model.scoring_history()
```

```python
print(scoring_hist.shape)
```

```python
scoring_hist
```

# Now looking into logloss metrics

```python
scoring_hist['training_logloss']
```

# Now creating the training logloss base data frame and then ranking it

```python
diff_df = scoring_hist['training_logloss'].diff()
diff_df.rank()
```

# Now creating the training auc base data frame and then ranking it

```python
diff_df = scoring_hist['training_auc'].diff()
diff_df.rank()
```
